```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('C:/Users/26919/Desktop/Survey/dataset/metr-la/METR-LA.csv', index_col = 0)
```

```python
print(data.head)
```

```python
data.describe()
```

```python
nan_num = data.isna().sum()
print(nan_num)
```

```python
#采用移动平均方法对数据进行平滑处理，同时使用填充方法对缺失值进行填充
smoothed_data = data.rolling(window=5).mean().bfill()
```

```python
print(smoothed_data.head)
```

```python
smoothed_data.describe()
```

```python
smoothed_data[['773869', '767541', '717447', '773062', '717590']].hist(bins=15, figsize=(10, 6))  # 设置直方图的箱数和图形大小
plt.xlabel('Traffic Flow')  # 设置x轴标签
plt.ylabel('Frequency')  # 设置y轴标签
plt.title('Histogram of Traffic Flow')  # 设置标题
plt.show()
```

```python
smoothed_data[['773869', '767541', '717447', '773062', '717590']][:1000].plot(figsize=(10, 6), label=['773869', '767541', '717447', '773062', '717590'])  # 设置图形大小
plt.xlabel('Time')  # 设置x轴标签
plt.ylabel('Traffic Flow')  # 设置y轴标签
plt.title('Traffic Flow Over Time')  # 设置标题
plt.legend(loc='upper right')  # 设置图例位置
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# 选择合适的参数值
eps = 0.1  # 领域半径
min_samples = 10  # 最小样本数
smoothed_data_transposed = smoothed_data.transpose()
# 初始化并拟合 DBSCAN 模型
dbscan = DBSCAN(eps=eps, min_samples=min_samples)
dbscan.fit(smoothed_data_transposed)

# 获取聚类结果
labels = dbscan.labels_

# 绘制聚类结果的散点图
plt.scatter(range(len(smoothed_data_transposed)), labels, c=labels, cmap='viridis')
plt.xlabel('Time')
plt.ylabel('Cluster Label')
plt.title('DBSCAN Clustering of Traffic Flow Data')
plt.show()
```

```python
# 创建散点图
plt.figure(figsize=(48, 24), dpi = 144)  # 设置图形大小

# 绘制散点图
plt.scatter(smoothed_data.index[:1000], smoothed_data['773869'][:1000], label='773869', s=50, c='blue')
plt.scatter(smoothed_data.index[:1000], smoothed_data['717590'][:1000], label='717590', s=50, c='red')
plt.scatter(smoothed_data.index[:1000], smoothed_data['773062'][:1000], label='773062', s=50, c='green')
plt.scatter(smoothed_data.index[:1000], smoothed_data['717447'][:1000], label='717447', s=50, c='purple')

# 设置标签
plt.xlabel('Time')
plt.ylabel('Traffic Flow')

# 设置标题
plt.title('Traffic Flow Scatter Plot')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```
